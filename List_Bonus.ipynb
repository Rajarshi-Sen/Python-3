```python
import random
import turtle

move=15    #5 #3 #15   #this is the distance by which the turtle moves
angle=20 #15 #this is the angle used to turn the turtle
direction=[] #remembers the last directions taken by the turtle
position=[]  #remembers the last positions held by the turtle

def turtleLetter(letter):
    if letter=='F':
        turtle.forward(move)
    elif letter=='A':
        turtle.forward(move)
    elif letter=='B':
        turtle.forward(move)
    elif letter=='+':
        turtle.left(angle)
    elif letter=='-':
        turtle.right(angle)
    elif letter=='[':
        direction.append(turtle.heading())
        position.append(turtle.pos())
    elif letter==']':
        turtle.penup()
        turtle.setheading(direction.pop())
        turtle.setposition(position.pop())
        turtle.pendown()
    elif letter=='0':
        turtle.color('darkgreen')
    elif letter=='1':
        turtle.color('green')
    elif letter=='2' or letter == '3':
        turtle.color('brown')

def growAndDraw(pattern,seed,times):
    turtle.speed(500)
    #turtle.left(90)
    while times>0:
        L2=[]
        index=0
        while index<len(seed):
            L2=L2+list(pattern(seed[index]))
            index=index+1
        seed=L2
        times=times-1
    turtleSpeak(seed)

def plantPattern1(x):
    if x=='F':
        return 'F','[','-','F',']','F','[','+','F',']','[','F',']'
    return x

def turtleSpeak(listOfLetters):
    #map(turtleLetter,listOfLetters)
    for element in listOfLetters:
        turtleLetter(element)
    return

# setting the initial position and direction:
def turtle_Set_Initial(x, y, t_dir):
    turtle.penup()
    turtle.setx(x)
    turtle.sety(y)
    turtle.setheading(t_dir)
    turtle.pendown()
    return

def plantPattern2(x):
    if x == 'F':
        return 'F','[','+','F',']','F','[','-','F',']'
    return x        

def plantPattern3(x):
    if x == 'F':
        return 'F','F'
    elif x == 'X':
        return 'F','[','+','X',']','F','[','-','X',']','+','X'
    return x


def coloredPattern(x):
    if x == 'F':
        return 'F','F'
    elif x == 'X':
        return '0','F','-','[','2','[','X',']','+','3','X',']','+','1','F','[','3','+','F','X',']','-','X'
    return x

def randomPattern(x):
    rand = random.uniform(0,100)
    if x == 'X':
        return 'F','[','+','X',']','F','[','-','X',']','+','X'
    elif x == 'F':
        if rand < 20:
            return 'F','F','+','F'
        else:
            return 'F','[','+','F',']'
    return x   

def flowerPattern(x):
    if x == 'X':
        return 'F', '-', '[', '[', 'X', ']', '+', 'X', ']', '+', 'F', '[', '+', 'F', 'X', ']', '-', 'X'
    elif x == 'F':
        return 'F','F'
    return x
    
```
